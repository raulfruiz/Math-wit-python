```
#La importación de librerías de toda la vida
import numpy as np
import matplotlib.pyplot as plt

def g(x):
  return x**2 #definición de la primera función (x^2)

def f(x):
  return np.sin(x) #definición de la segunda, sen(x)

x = np.linspace(-10,10, num=1000) #creación de vector x (-10,10) con mil puntos incluídos

a=f(x)
b=g(x)
f_o_g = f(g(x)) #composición de la función f(x) en g(x)
g_o_f = g(f(x)) #lo mismo pero al revés

fig, ax = plt.subplots()

ax.plot(x,a, color='y')#x^2 en amarillo
ax.plot(x,b, color='c') #sen(x) en cyan
ax.plot(x,f_o_g, color='b') #gráfica de f compuesto g, azul
ax.plot(x,g_o_f, color='g') #gráfica de g compuesto f, verde 

"Las 4 líneas anteriores se pueden comentar una por una para analizar mejor cada caso"

ax.grid()
ax.axhline(y=0, color='r')
ax.axvline(x=0, color='r')
```
