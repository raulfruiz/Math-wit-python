## Dependencias

```
from matplotlib import cm # Para manejar colores de las curvas
import numpy as np #Para las operaciones bellas
import matplotlib.pyplot as plt #para ser más cool graficando
```

## Gráfica en 3 dimensiones

```

def f(x,y):
  return np.sin(x)+2*np.cos(y)


res = 100

x = np.linspace(-4, 4, num= res)
y = np.linspace(-4, 4, num=res)

x, y = np.meshgrid(x, y) #retorna la combinación de pares coordenados formados por los ejes X y Y

z=f(x,y) 

# Gráficar la superficie
fig, ax = plt.subplots(subplot_kw={"projection": "3d"}) #Esto para que grafique una superficie en 3 ejes, y no una curva en 2 ejes
surf = ax.plot_surface(x,y,z, cmap=cm.cool, linewidth=0 )#Declaración de superficie a graficar
#cmap es para indicar con colores los puntos más altos y bajos en el eje Z.


fig.colorbar(surf) #y colorbar() es para crear una barra que indica los valores de z asignados a cada color

```

## Curvas de nivel

```
#Graficando curvas de nivel (2D) en vez de superficies (3D)
#Las curvas de nivel son planos que, a pesar de estar en 2D, representan superficies (que están en 3D)

fig2,ax2 = plt.subplots()
level_map = np.linspace(np.min(z), np.max(z),num=res)#Creación de vector para obtener los valores mínimos y máximos de la superficie en Z 

#Ahora se van a graficar las curvas de nivel
cp = ax2.contour(x, y, z, levels=level_map,cmap=cm.cool) #contour() traza las curvas de nivel. La documentación de esta funnción
# en matplotlib está bastante completa.
fig2.colorbar(cp) #la barrita de la derecha para indicar altura de z
ax2.set_title('Curvas de nivel') #ponerle título a la cuestión
plt.show() #enjoy
```
